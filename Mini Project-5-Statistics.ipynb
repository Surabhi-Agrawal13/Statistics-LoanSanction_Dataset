### Q-18. Check outliers in Loan amount and credit history column and replace them with mean.

```python
DF['LoanAmount'] = pd.to_numeric(DF['LoanAmount'],errors='coerce')
DF['Credit_History'] = pd.to_numeric(DF['Credit_History'],errors='coerce')

loan_amount_mean = DF['LoanAmount'].mean()
credit_history_mean = DF['Credit_History'].mean()

def replace_outliers(column):
    z_scores = (column - column.mean()) / column.std()
    outliers = (z_scores > 3) | (z_scores < -3)
    column[outliers] = column.mean()

replace_outliers(DF['LoanAmount'])
replace_outliers(DF['Credit_History'])

print(DF)
```
